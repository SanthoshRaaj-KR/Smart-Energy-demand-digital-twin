# Task
Generate heatmaps for the correlation matrices of the numerical columns for each of the four individual state DataFrames (`df_bihar`, `df_nrup`, `df_punjab`, `df_west_bengal`). Each heatmap should be clearly displayed, and the 'State' column must be dropped before calculating the correlations for each state.

```python
import seaborn as sns
import matplotlib.pyplot as plt

print("Seaborn and Matplotlib imported successfully.")
```

```python
state_dataframes_list = {
    'Bihar': df_bihar,
    'NR UP': df_nrup,
    'West Bengal': df_west_bengal,
    'SR Karnataka': df_karnataka
}

for state_name, df_state in state_dataframes_list.items():

    # Ensure Date is datetime
    df_state['Date'] = pd.to_datetime(df_state['Date'])

    # Extract day, month, year
    df_state['dd'] = df_state['Date'].dt.day
    df_state['mm'] = df_state['Date'].dt.month
    df_state['yyyy'] = df_state['Date'].dt.year

    # Drop non-numeric columns for correlation
    df_state_numeric = df_state.drop(columns=['State', 'Date'])

    # Calculate correlation matrix
    correlation_matrix = df_state_numeric.corr()

    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap='coolwarm',
        fmt='.2f',
        linewidths=.5
    )
    plt.title(f'Correlation Matrix for {state_name}', fontsize=16)
    plt.show()
```

```python
print(df_state.columns)
```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import r2_score, mean_squared_error

# -----------------------------
# 1️⃣ Feature Engineering (per state)
# -----------------------------

for state_name, df_state in state_dataframes_list.items():

    df_state = df_state.sort_values('Date').reset_index(drop=True)

    df_state = df_state.drop(columns=['Drawl_Schedule', 'Gap', 'om_precip'])

    # Encode year
    le_year = LabelEncoder()
    df_state['yyyy'] = le_year.fit_transform(df_state['yyyy'])

    # Cyclical encoding
    df_state['dd_sin'] = np.sin(2 * np.pi * df_state['dd'] / 31)
    df_state['dd_cos'] = np.cos(2 * np.pi * df_state['dd'] / 31)
    df_state['mm_sin'] = np.sin(2 * np.pi * df_state['mm'] / 12)
    df_state['mm_cos'] = np.cos(2 * np.pi * df_state['mm'] / 12)

    df_state = df_state.drop(columns=['dd', 'mm'])

    # Lag features
    for lag in [1, 7, 14, 28]:
        df_state[f'lag_{lag}'] = df_state['Actual_Drawl'].shift(lag)

    # Rolling means
    for window in [1, 7, 14, 28]:
        df_state[f'roll_mean_{window}'] = (
            df_state['Actual_Drawl']
            .rolling(window=window)
            .mean()
            .shift(1)
        )

    # Rolling std
    df_state['roll_std_7'] = (
        df_state['Actual_Drawl']
        .rolling(window=7)
        .std()
        .shift(1)
)

    state_dataframes_list[state_name] = df_state
```

```python
print(df_state.columns)
print(df_state.shape)
display(df_state.head())
display(df_state.tail())
```

```python
# -----------------------------
# 2️⃣ Merge All States (FIXED)
# -----------------------------

cleaned_states = []

for df_state in state_dataframes_list.values():
    cleaned_states.append(df_state)

full_df = pd.concat(cleaned_states, ignore_index=True)

# Encode state
le_state = LabelEncoder()
full_df['State'] = le_state.fit_transform(full_df['State'])

# Sort globally by date
full_df = full_df.sort_values('Date').reset_index(drop=True)
```

```python
print(full_df.columns)
# print(full_df.shape)
# display(full_df.head())
```

```python
# -----------------------------
# 3️⃣ Train / Validation Split (BEFORE scaling)
# -----------------------------

max_date = full_df['Date'].max()
cutoff_date = max_date - pd.DateOffset(months=24)

train_df = full_df[full_df['Date'] <= cutoff_date]
val_df = full_df[full_df['Date'] > cutoff_date]

target = 'Actual_Drawl'

X_train = train_df.drop(columns=[target, 'Date'])
y_train = train_df[target]

X_val = val_df.drop(columns=[target, 'Date'])
y_val = val_df[target]
```

```python
# -----------------------------
# 4️⃣ Scaling (FIT ONLY ON TRAIN)
# -----------------------------

climatic_cols = ['om_temp_mean', 'nasa_solar', 'om_dewpoint', 'om_wind_gusts']
lag_roll_cols = [col for col in X_train.columns if 'lag_' in col or 'roll_' in col]

# Scale climatic features
scaler_climate = StandardScaler()
X_train[climatic_cols] = scaler_climate.fit_transform(X_train[climatic_cols])
X_val[climatic_cols] = scaler_climate.transform(X_val[climatic_cols])

# Scale lag + rolling features
scaler_lagroll = StandardScaler()
X_train[lag_roll_cols] = scaler_lagroll.fit_transform(X_train[lag_roll_cols])
X_val[lag_roll_cols] = scaler_lagroll.transform(X_val[lag_roll_cols])

# Scale target separately
scaler_target = StandardScaler()
y_train = scaler_target.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_val = scaler_target.transform(y_val.values.reshape(-1, 1)).ravel()
```

```python
# -----------------------------
# 5️⃣ LightGBM Training
# -----------------------------

lgb_train = lgb.Dataset(
    X_train,
    label=y_train,
    categorical_feature=['State']
)

lgb_val = lgb.Dataset(
    X_val,
    label=y_val,
    categorical_feature=['State'],
    reference=lgb_train
)

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'verbose': -1
}

model = lgb.train(
    params,
    lgb_train,
    valid_sets=[lgb_val],
    num_boost_round=1000
)

val_preds = model.predict(X_val)

r2 = r2_score(y_val, val_preds)
rmse = np.sqrt(mean_squared_error(y_val, val_preds))

print("\nValidation Results:")
print(f"R2 Score  : {r2:.4f}")
print(f"RMSE      : {rmse:.4f}")
```
